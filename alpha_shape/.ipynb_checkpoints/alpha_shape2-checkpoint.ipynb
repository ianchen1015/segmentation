```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Delaunay
#https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.spatial.Delaunay.html

img = np.array(cv2.imread("./input/1.jpg", 0))
print (img.shape)
pointNumber = 0
x = []
y = []
point_lst = []
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        if img[i][j] > 150:
            pointNumber += 1
            x.append(j)
            y.append(i)
            point_lst.append([i, j])
print(pointNumber)
points = np.array(point_lst)

plt.figure(figsize=(20,10))
plt.plot(x,y,'.', color='#f16824')
```

```python
tri = Delaunay(points)

plt.figure(figsize=(20,25))
plt.triplot(points[:,0], points[:,1], tri.simplices.copy())
#plt.plot(points[:,0], points[:,1], 'o')
plt.show()
```

```python
print(tri)
#print(points[tri.simplices])

def d(p1, p2):
    return (abs(p1[0] - p2[0]) ** 2 + abs(p1[1] - p2[1]) ** 2) ** 0.5

def draw_edge_list(output_edges):
    plt.figure(figsize=(5,10))
    for edge in output_edges:
        plt.plot([edge[0][0],edge[1][0]], [edge[0][1],edge[1][1]], '-')
    plt.show()
    
def alpha_shape(alpha):
    output_edges = []

    for tri_points in points[tri.simplices]:
        #print(tri_points)
        p1 = tri_points[0]
        p2 = tri_points[1]
        p3 = tri_points[2]
        d1 = d(p1, p2)
        d2 = d(p2, p3)
        d3 = d(p1, p3)

        if d1 < alpha and d2 < alpha and d3 < alpha:
            pass
        elif d1 > alpha and d2 > alpha and d3 > alpha:
            pass
        else:
            #print(min(d1, d2, d3))
            min_d = min(d1, d2, d3)
            if min_d == d1:
                output_edges.append([p1, p2])
            if min_d == d2:
                output_edges.append([p2, p3])
            if min_d == d3:
                output_edges.append([p1, p3])
    #print(output_edges)
    draw_edge_list(output_edges)

for i in range(2,10):
    print('alpha = ',i)
    alpha_shape(i)
```
